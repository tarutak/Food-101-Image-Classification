# Inference from the trained model

### Here we demostrate how to take our best trained classifer and use it to predict to predict a food dish. 

#### Importing Libraries

```python
from fastai import *
from fastai.vision import *
```

#### Setting path for the model, i.e., the .pkl file and Images we want to predict. 

```python
#set this path according to your local machine
path=Path('/home/jupyter/.fastai/data/food-101/images')
```

#### Creating a learn object from our saved model

```python
learn=load_learner(path,file="food101_ResNet50_87%.pkl")
```

#### loading an image to get prediction

```python
# use this img variable to use your own image
img = open_image(path/'apple_pie'/'420768.jpg')
img
```

#### Prediction

```python
pred_class,pred_idx,outputs = learn.predict(img)
pred_class
```
